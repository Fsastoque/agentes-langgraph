```python
from google import genai
from google.genai import types

# 1. Inicializa el cliente (busca automáticamente la variable de entorno GEMINI_API_KEY)
client = genai.Client()

# 2. Sube tu archivo PDF directamente al almacenamiento de Gemini
# Nota: Reemplaza "tu_archivo.pdf" con la ruta real de tu PDF
archivo_pdf = client.files.upload(
    file="/home/fsastoque/desarrollo/langgraph/agentes-langgraph/Mobile_Site_Speed_Playbook.pdf",    
)

print("Archivo PDF subido con ID:", archivo_pdf.name)

mi_almacen_rag = client.caches.create(
    model="gemini-2.5-flash",
    config=types.CreateCachedContentConfig(
        contents=[archivo_pdf],
        # Opcional: TTL (Tiempo de vida). Puedes definir cuánto tiempo persistirá
        # ttl="86400s" # 24 horas, o puedes omitirlo para manejo estándar
    )
)

response = client.models.generate_content(
    model='gemini-2.5-flash',
    contents="¿Cuáles son los puntos clave explicados en el documento?",
    config=types.GenerateContentConfig(
        cached_content=mi_almacen_rag.name
    )
)

print("\nRespuesta de Gemini:")
print(response.text)
```
