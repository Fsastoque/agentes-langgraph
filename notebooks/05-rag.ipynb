```python
from google import genai
from google.genai import types

# 1. Inicializa el cliente (busca automáticamente la variable de entorno GEMINI_API_KEY)
client = genai.Client()

# 2. Sube tu archivo PDF directamente al almacenamiento de Gemini
# Nota: Reemplaza "tu_archivo.pdf" con la ruta real de tu PDF
archivo_pdf = client.files.upload(
    file="/home/fsastoque/desarrollo/langgraph/agentes-langgraph/Mobile_Site_Speed_Playbook.pdf",    
)

print("Archivo PDF subido con ID:", archivo_pdf.name)


response = client.models.generate_content(
    model='gemini-2.5-flash',
    contents=[archivo_pdf, "¿Cuáles son los puntos clave explicados en el documento?"]
)

print("\nRespuesta de Gemini:")
print(response.text)
```

```python
from langchain.chat_models import init_chat_model

llm = init_chat_model("openai:gpt-4o", temperature=0)
```

```python
file_search_tool = {
    "type": "file_search",
    "vector_store_ids": ["vs_68cf0f0255e481919cd3be25b96c5080"],#Id del vector store que contiene los embeddings de tu PDF subido en OpenAI
}
llm_with_tools = llm.bind_tools([file_search_tool])
```

```python
response = llm_with_tools.invoke("Como mejoro el rendimiento de mi website?")
response.text
```
