```python
from langchain_core.messages import AIMessage

ai_message = AIMessage(content="Hello, how can I help you?")
ai_message.text
```

```python
from langchain_core.messages import HumanMessage

human_message = HumanMessage(content="What is the capital of France?")
human_message.text
```

```python
from langchain_core.runnables import history
history = [ai_message, human_message]
for m in history:
    m.pretty_print()
```

```python
history + [AIMessage(content="The capital of France is Paris.")]
```
